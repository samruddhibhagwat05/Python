```python
import numpy as np
data = {
    'A':[1,2,np.nan,4," ",np.nan],
    'B':[np.nan,2,3," ",np.nan,np.nan],
    'C':[1,2,3," ",4,np.nan],
    'D':[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]
}
df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dropna() ## drop all the rows with null values
```

```python
df.dropna(axis=1) ## drops all the columns which have null values present
```

```python
df.dropna(how='all') # this will remove row in which all values are blank. row 5 is removed
```

```python
df.dropna(axis=1,how='all') #remove columns in which each value is nan. column D is removed
```

```python
df.dropna(subset=['B']) #remove row where B column has nan value
```

```python
import pandas as pd 
import numpy as np

data = {
    'A':[1,2,np.nan,4,5],
    'B':[np.nan,8,9,np.nan,5],
    'C':[1,7,3,9,5]
   # 'D':['A','B','C','D','E']
}

df = pd.DataFrame(data)
df
```

```python
df.replace(np.nan,4) ## all the null values is replaced by value 4
```

```python
## Forward Fill
df.ffill() ## it fills the null value with its back value
```

```python
## Backward fill
df.bfill() ## it fills its null value with forward value
```

```python
df.fillna(df.mean()) ## fills all the null values with mean
```

## Grouping and Aggregation

```python
# Get value counts of lunch types by gender
df.groupby(['gender', 'lunch']).size()
```

```python
data = {
    'EmployeeID':[1,2,3,4,5,6,7,8,9],
    'Name':['Shahzad','Dheeraj','Kishore','Praveen','Sandeep','Deepak','Srikant',
           'Anandh','Sai'],
    'Department':['HR','Finance','HR','Finance','IT','IT','HR','Finance','IT'],
    'Salary':[60000,75000,62000,80000,90000,85000,61000,78000,92000],
    'Experience (Years)':[3,5,2,8,4,6,2,7,5]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby("Department")["Salary"].sum()
```

```python
df.groupby("Department")["Salary"].agg(["max","sum","mean"]) ## grouping multiple functions at a time
```

```python
df.groupby(["Department","Experience (Years)"])["Salary"].sum()
```

## Combining Data

```python
data1 = {'ID':[1,2,3,4],
        'Name':['alka','Kishore','Sai','Shamali']}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {
    'ID':[1,2,4,5],
    'Age':[25,30,35,28]
}
df2 = pd.DataFrame(data2)
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index = True)
```

```python
pd.concat([df1,df2],axis = 1)
```

```python
pd.merge(df1,df2,on = 'ID',how = 'inner')
```

```python
pd.merge(df1,df2,on='ID',how='right')
```
